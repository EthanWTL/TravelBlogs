```python
import os
import requests
```

```python
# Access the API key from the .env file
unwrangle_api_key = os.getenv('UNWRANGLE_API_KEY')
```

```python
# Define the API endpoint and parameters
url = "https://data.unwrangle.com/api/getter/"
params = {
    'platform': 'yelp_reviews',
    'biz_id': 'yTynZugQRl5GVVFAIEAtfQ',
    'page': 2,
    'api_key': unwrangle_api_key
}

# Make the GET request to the API
response = requests.get(url, params=params)

# Check if the request was successful
if response.status_code == 200:
    # Parse the JSON response (if the API returns JSON)
    data = response.json()
    print(data)
else:
    print(f"Error: {response.status_code}")
    print(response.text)
```
