```python
import requests
from bs4 import BeautifulSoup
```

```python
URL = "https://www.sincerelyreneze.com/blog/visiting-my-dream-city-chicago"
r = requests.get(URL)
soup = BeautifulSoup(r.content, 'html5lib')
print(soup.prettify())
```

```python
materials=[]  # a list to store quotes

article = soup.find('article', attrs = {'id':'post-642998ba95dda1251b1d4cd2'})

for block in article.findAll('div', attrs = {'class':'sqs-html-content'}):
    title_2 = block.findAll('h2')
    title_3 = block.findAll('h3')
    mains = block.findAll('p')
    
    if(title_2 != []):
        for title in title_2:
            print(title.text)
            
    if(title_3 != []):
        for title in title_3:
            print(title.text)
            
    if(mains != []):
        for main in mains:
            print(main.text)
    
```
